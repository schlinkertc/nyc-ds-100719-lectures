```python
import numpy as np
import pandas as pd
np.random.seed(0)
df = pd.DataFrame({'state': ['CA', 'WA', 'CO', 'AZ'] * 3,
                   'office_id': list(range(1, 7)) * 2,
                   'sales': [np.random.randint(100000, 999999)
                             for _ in range(12)]})
```

```python
df
```

## Show sales as a percentage of total sales in their state


```python
df[df['state']=='CA'].groupby('office_id').sum()
```

```python
total_cal_sales = df[df['state']=='CA'].sum()[2]
print(total_cal_sales)
df['state'].unique()
```

```python
# group by state and office id 

state_office = df.groupby(['state','office_id']).sum()
state_office
```

```python
# group the group: level=0 refers to the first index of a multi-index dataframe
state_office.groupby(level=0).sum()
```

```python
# apply a percentile lambda function to the grouped df 

state_pcts = state_office.groupby(level=0).apply(lambda x: 100*x/float(x.sum()))
```

```python
state_pcts
```
